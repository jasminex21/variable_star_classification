```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
```

```python
data = {
    'RF': [1.000000, 0.915429],
    'XGB': [0.915429, 1.000000]
}

index = ['RF', 'XGB']

# Create the DataFrame
crosscorr = pd.DataFrame(data, index=index)
crosscorr
```

```python
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(crosscorr, annot=True, cmap='coolwarm', fmt=".3f", vmin=0, vmax=1)
plt.title('Cross-Correlation Matrix between RF and XGB Models', fontsize=16, fontweight="bold")
plt.savefig('cross-correlation_matrix.png')
plt.show()
```
